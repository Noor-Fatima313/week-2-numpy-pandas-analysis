```python
import numpy as np
import pandas as pd
```

# Dataset Loading

In this section, the Titanic dataset is loaded using Pandas. We then inspect its structure by viewing sample records, dimensions, column names, and data types.

```python
# Load dataset
df = pd.read_csv('/content/titanic.csv')

# Display first 10 rows
print("First 10 Rows")
display(df.head(10))

# Display last 5 rows
print("Last 5 Rows")
display(df.tail())

# Dataset dimensions
print("Dataset Shape:", df.shape)

# Column names
print("\nColumn Names:")
print(df.columns.tolist())

# Data types
print("\nData Types:")
print(df.dtypes)
```

# Data Exploration

The dataset is explored to identify missing values, duplicate records, descriptive statistics, and feature types.

```python
# Missing values
print("Missing Values")
print(df.isnull().sum())

# Duplicate records
print("\nDuplicate Records")
print(df.duplicated().sum())

# Descriptive statistics
print("\nDescriptive Statistics")
display(df.describe())

# Categorical columns
print("\nCategorical Features")
print(df.select_dtypes(include='object').columns.tolist())

# Numerical columns
print("\nNumerical Features")
print(df.select_dtypes(include=np.number).columns.tolist())
```

# Data Cleaning

This section handles missing values, removes duplicate records, renames columns, and converts data types where required.

```python
# Fill missing Age values with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill missing Embarked values with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Replace missing Cabin values
df['Cabin'] = df['Cabin'].fillna('Unknown')

# Remove duplicates
df.drop_duplicates(inplace=True)

# Rename column
df.rename(columns={'Pclass':'Passenger_Class'}, inplace=True)

# Convert Survived to Boolean
df['Survived'] = df['Survived'].astype(bool)

print("Missing Values After Cleaning")
print(df.isnull().sum())
```

# Data Manipulation

Various Pandas operations are performed to analyze and transform the dataset.

```python
# 1. Filter passengers older than 30
print("Passengers Older Than 30")
display(df[df['Age'] > 30])

# 2. Sort by Fare
print("Highest Fare")
display(df.sort_values(by='Fare', ascending=False).head())

# 3. Group by Gender
print("Average Fare by Gender")
display(df.groupby('Sex')['Fare'].mean())

# 4. Aggregate Age by Passenger Class
print("Age Statistics by Passenger Class")
display(
    df.groupby('Passenger_Class')['Age'].agg(['mean', 'max', 'min'])
)

# 5. Create Family Size column
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

print("Dataset with New Column")
display(df.head())
```

```python
display(df[['SibSp', 'Parch', 'FamilySize']].head(10))
```

# Export Processed Dataset

The cleaned dataset is exported as a CSV file for future use.

```python
# Save cleaned dataset
df.to_csv('cleaned_titanic.csv', index=False)

print("Cleaned dataset exported successfully.")
```
